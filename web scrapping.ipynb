```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.theasiacollective.com/best-hotels-india/"
```

```python
Website_Allow=requests.get(url)
```

```python
Website_Allow
```

```python
soup=BeautifulSoup(Website_Allow.text)
```

```python
soup
```

```python
besthotels=soup.select("strong a")
```

```python
besthotels
```

```python
besthotels_list=[]
```

```python
for i in besthotels:
    besthotels_list.append(i.text)
```

```python
besthotels_list
```

```python
address=soup.select("h4+ p a")
```

```python
address
```

```python
address_list=[]
```

```python
for z in address:
    address_list.append(z.text)
```

```python
address_list
```

```python
instaaccount=soup.select("p+ p strong+ a")
```

```python
instaaccount
```

```python
instaaccount_list=[]
```

```python
for j in instaaccount:
    instaaccount_list.append(j.text)
```

```python
instaaccount_list
```
